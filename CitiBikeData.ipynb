```python
import pandas as pd
import os
import glob
```

```python
# Change directory
results_loc = "./"#input("Enter the path for your download directory: ")
try:
    os.chdir(results_loc)
except FileNotFoundError:
    print("Please change the directory to the location of your results files")

# List all files and directories in current directory
files = glob.glob((results_loc + "/*.csv").split("/")[-1])
print(files)
```

```python
df = pd.DataFrame(columns=['tripduration',
'starttime',
'stoptime',
'start station id',
'start station name',
'start station latitude',
'start station longitude',
'end station id',
'end station name',
'end station latitude',
'end station longitude',
'bikeid',
'usertype',
'birth year',
'gender'])
df
```

```python
file_count = 0
for item in files:

    try:
        print(f"Parsing {item}")
        quarter = item.split('-')[1]
        xl = pd.read_csv(item, names=['tripduration','starttime','stoptime','start station id','start station name',
                                      'start station latitude', 'start station longitude', 'end station id',
                                      'end station name', 'end station latitude', 'end station longitude', 'bikeid',
                                      'usertype', 'birth year', 'gender'], header=None, skiprows=[0])

        date = item.split('-')[1]
        xl["year"] = date[0:4]
        xl["month"] = date[4:7]
        print(f'{date[0:4]} and {date[4:7]}')
        df = df.append(xl, ignore_index=True, sort=False)
        file_count+=1
    except XLRDError:
        print("Close any open files and try again")
        
print(f"Parsed {file_count} files into dataframe df")
```
